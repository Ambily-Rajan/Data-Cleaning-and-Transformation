```python
# Filter rows where 'Number_of_Employees' column is NaN
result_employees_nan = df[df['Number_of_Employees'].isna()]

# Iterate over the filtered rows and apply the specified conditions
for index, row in result_employees_nan.iterrows():
    if pd.notna(row['Location']) and 'Employees' in str(row['Location']):
        df.at[index, 'Number_of_Employees'], df.at[index, 'Location'] = row['Location'], np.nan
    elif pd.notna(row['Type']) and 'Employees' in str(row['Type']):
        df.at[index, 'Number_of_Employees'], df.at[index, 'Type'] = row['Type'], np.nan
    elif pd.notna(row['Age']) and 'Employees' in str(row['Age']):
        df.at[index, 'Number_of_Employees'], df.at[index, 'Age'] = row['Age'], np.nan
    elif pd.notna(row['Expertise']) and 'Employees' in str(row['Expertise']):
        df.at[index, 'Number_of_Employees'], df.at[index, 'Expertise'] = row['Expertise'], np.nan

```
